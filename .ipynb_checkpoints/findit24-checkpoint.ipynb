## Outlier Handling

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# pendapatan
sns.boxplot(x=df_train1["pendapatan"], ax=axes[0])
axes[0].set_title("Pendapatan")
axes[0].set_xlabel("Pendapatan")

# belanja_daging
sns.boxplot(x=df_train1["belanja_daging"], ax=axes[1])
axes[1].set_title("Belanja Daging")
axes[1].set_xlabel("Belanja Daging")

# pembelian_web
sns.boxplot(x=df_train1["pembelian_web"], ax=axes[2])
axes[2].set_title("Pembelian Web")
axes[2].set_xlabel("Pembelian Web")

plt.show()
```

```python
# count outlier
def detect_outliers_tukey(df, k=1.5):
    outliers_count = {}
    for column in df.columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - k * IQR
        upper_bound = Q3 + k * IQR
        outliers_count[column] = df[column][(df[column] < lower_bound) | (df[column] > upper_bound)].count()
    return outliers_count

outliers_count = detect_outliers_tukey(df_train1)

for column, count in outliers_count.items():
    print(f"Number of outliers in {column}: {count}")
```

```python
# Menghitung Q1, Q3, dan IQR
Q1 = df_train1['pendapatan'].quantile(0.25)
Q3 = df_train1['pendapatan'].quantile(0.75)
IQR = Q3 - Q1
# Menentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Mendeteksi outlier
outliers = df_train1[(df_train1['pendapatan'] < lower_bound) | (df_train1['pendapatan'] > upper_bound)]
```

```python
outliers
```

```python
# remove outlier
df_train1 = df_train1[(df_train1["pendapatan"] >= lower_bound) & (df_train1["pendapatan"] <= upper_bound)]
print("Dataset setelah menghapus outlier:\n", df_train1)
```
